```python
import pandas as pd
import numpy as np
import requests
from io import StringIO
```

```python
url="https://raw.githubusercontent.com/dineshpiyasamara/LaptopPricePredictor/master/model%20building/laptop_price.csv"
response=requests.get(url)
```

```python
if response.status_code==200:
    data = pd.read_csv(StringIO(response.text))
else:
    print("failed to grab from github")
```

```python
data.head()
```

```python
data["Ram"]=data["Ram"].str.replace("GB","").astype('int32')
```

```python
data.info()
```

```python
data['Weight']=data['Weight'].str.replace('kg','').astype('float')
```

```python
data.info()
data.head()
```

```python
data.drop('laptop_ID',axis=1)
```

```python
def func(inpt):
    if inpt == 'Samsung' or inpt == 'Razer' or inpt == 'Mediacom' or inpt == 'Microsoft'or inpt == 'Xiaomi'or inpt == 'Vero'or inpt == 'Chuwi'or inpt =='Google'or inpt == 'Fujitsu' or inpt == 'LG'or inpt == 'Huawei':
        return 'other'
    else:
        return inpt
    
data['Company']=data['Company'].apply(func)
```

```python
data['Product'].value_counts()
```

```python
data.drop('Product',axis=1)
```

```python
data['ScreenResolution'].value_counts
```

```python
data['Touchscreen'] =data['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
data['IPS panel']   =data['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
data['Full HD']     =data['ScreenResolution'].apply(lambda x:1 if 'Full' in x else 0)
```

```python
data.drop('ScreenResolution',axis=1)
```

```python
data['cpu_name']=data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
data.drop('Cpu',axis=1)
```

```python
data['cpu_name'].value_counts()
```

```python
def set_processor(name):
    if name=='Intel Core i7' or name=='Intel Core i5' or name=='Intel Core i3' or name=='Intel Celeron Dual' or name=='Intel Pentium Quad' or name=='Intel Core M':
        return name
    else:
        if name.split()[0]=='AMD':
            return 'AMD'
        
        else:
            return 'other'
        
data['cpu_name']=data['cpu_name'].apply(set_processor)
```

```python
def func(name):
    y=name.split()[0:1]
    x="".join(y)
    return x
data['Gpu name']=data['Gpu'].apply(func)
```

```python
def func(x):
    if x=='Windows 10' or x=="Windows 7":
        return 'Windows'
    elif (x=="macOS" or x=='Mac OS X'):
        return 'Mac'
    elif x=="Linux":
        return x
    else:
        return 'other'

data['OpSys'] = data['OpSys'].apply(func)
```

```python
data= data.drop(['laptop_ID','Inches','Product','Cpu','Gpu'], axis=1)
```

```python
data=data.drop('ScreenResolution',axis=1)
```

```python
data.head(4)
```

```python
#using one hot encoding
#new columns will be created for 'company','typename','cpu_name','Gpu name'
data=pd.get_dummies(data)
```

```python
data.head(2)
```

```python
data.shape
```

```python
X= data.drop('Price_euros',axis=1)
y=data['Price_euros']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.25)
```

```python
X_train.shape, X_test.shape
```

```python
#to train on different models
def model_accuracy(model):
    model.fit(X_train, y_train)
    accuracy= model.score(X_test, y_test)
    print('accuracy of'+str(model)+"is: "+str(accuracy))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model_accuracy(lr)
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
model_accuracy(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_accuracy(dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
model_accuracy(rf)
```

```python
#using grid search cv to find the best hyper absparameter
```

```python
from sklearn.model_selection import GridSearchCV
parameters={'n_estimators':[1,5,10,50,100],'criterion':['squared_error','absolute_error','poisson']}
gridObj= GridSearchCV(estimator=rf, param_grid=parameters)
gridFit= gridObj.fit(X_train, y_train)
bestModel=gridFit.best_estimator_
bestModel
```

```python
bestModel.score(X_test, y_test)
```
